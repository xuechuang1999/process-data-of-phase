```python
def read_csv_as_columns(file_path):
    """
    Reads a CSV file with tab-separated values and returns each column as a list.
    Assumes the first row is a comment line and skips it.
    
    Parameters:
        file_path (str): Path to the CSV file.

    Returns:
        list: A list of four lists, each containing the data of one column.
    """
    phase, pbe, pbesol, hse = [], [], [], []

    # Open the file and read the contents
    with open(file_path, 'r') as file:
        reader = csv.reader(file, delimiter=',')
        
        # Skip the first row (comment line)
        next(reader)
        
        # Read the remaining rows and append data to respective lists
        for row in reader:
            if len(row) == 4:  # Ensure the row has exactly 4 columns
                phase.append(row[0])
                pbe.append(row[1])
                pbesol.append(row[2])
                hse.append(row[3])
            else:
                print(f"Skipping row due to incorrect format: {row}")

    return phase, pbe, pbesol, hse

# Example usage
file_path = 'different_xc_energy_data.csv'  # Replace with your actual file path
phase, pbe, pbesol, hse = read_csv_as_columns(file_path)

print("Column 1:", phase)
print("Column 2:", pbe)
print("Column 3:", pbesol)
print("Column 4:", hse)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib



matplotlib.rcParams['font.family'] = 'Arial'

def plot_comparison_bar_chart(csv_file):
    """
    This function reads a CSV file and plots a grouped bar chart comparing values
    from the second to fourth columns across different phases.
    
    Parameters:
    csv_file (str): Path to the input CSV file.
    """
    # Read the CSV file
    data = pd.read_csv(csv_file)
    
    # Extract data columns
    phases = data['phase']  # First column as phase labels
    pbe = data['PBE']  # Second column (PBE)
    pbesol = data['PBESol']  # Third column (PBESol)
    hse = data['HSE']  # Fourth column (HSE)
    
    # Set the position for each group of bars
    x = np.arange(len(phases))  # Position of each group of bars
    width = 0.2  # Width of each bar
    
    # Create a figure and axis for the plot
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # Plot each set of bars with different colors
    ax.bar(x - width, pbe, width, label='PBE', color='tab:orange')
    ax.bar(x, pbesol, width, label='PBESol', color='tab:blue')
    ax.bar(x + width, hse, width, label='HSE', color='tab:green')
    
    # Set axis labels and title
    ax.set_xlabel(r'The different phase of $\text{ZrO}_2$', fontsize=12, labelpad=15)
    ax.set_ylabel('Relative Energy (Ev)', fontsize=12)
    ax.set_title('The energy with respect to P21_c phase', fontsize=14)
    
    # Set x-axis ticks and labels (phase names)
    ax.set_xticks(x)
    ax.set_xticklabels(phases, rotation=0, ha='center')
    
    # Add legend to the plot
    # ax.legend(title="Functional", fontsize=10)
    ax.legend(fontsize=10)
    
    # Adjust layout to prevent clipping of labels
    plt.tight_layout()
    
    # Show the plot
    plt.show()

# Example usage
csv_file = 'different_xc_energy_data.csv'  # Replace with your CSV file path
plot_comparison_bar_chart(csv_file)
```
